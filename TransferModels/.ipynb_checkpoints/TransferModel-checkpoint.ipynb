```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# --- Параметры задачи ---
x0, x1 = 0, 10
t0, t1 = 0, 100
f = 10.1
```

```python
# Функции V(x) и начальное/граничное условия
V = x + 1
U0 = lambda x: np.maximum(0, (4 - x) * (x - 9))
U_left = lambda t: t * (t - 20) * (t - 50) / 20000
```

```python
# Шаги сетки
dt_dx = [[0.05, 0.5], [0.025, 0.25], [0.01, 0.1], [0.005, 0.05], [0.001, 0.01]] 
```

```python
x = np.arange(x0, x1 + dt_dx[0][0], dt_dx[0][0])
t = np.arange(t0, t1 + dt_dx[0][1], dt_dx[0][1])
nx = len(x)
nt = len(t)
```

```python
def rmse(a, b):
    return np.sqrt(np.mean((a - b)**2))
```

```python
def U_analytic_scalar(tt, xx):
    # tt >= 0, xx >= 0
    # вычисляем x0
    x0 = -1.0 + (xx + 1.0) * np.exp(-tt)
    if x0 >= 0.0:
        return U0(x0) + f * tt
    else:
        # момент пересечения с x=0
        tb = tt - np.log(xx + 1.0)
        # tb будет >= 0 в этом случае
        return U_left(tb) + f * (tt - tb)
```

```python
# Векторизуем для удобства (работает быстрее при построении сетки)
U_analytic = np.vectorize(U_analytic_scalar)

# ---------- Выбор моментов времени для профилей ----------
# используем те же t как в численном коде и 5 профилей как раньше
indices = np.linspace(0, nt - 1, 5, dtype=int)  # 5 равномерных индексов от 0 до nt-1
t_snapshots = t[indices]

# ---------- Построение графиков ----------
plt.figure(figsize=(9, 6))
for tt in t_snapshots:
    U_vals = U_analytic(tt, x)
    plt.plot(x, U_vals, label=f"t = {tt:.1f}")

plt.xlabel('x')
plt.ylabel('U(t, x)')
plt.title('Аналитическое решение (метод характеристик)')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

**Вывод схем численного решения**

$\frac{∂U}{∂t}+V(x)\frac{∂U}{∂x}=f(x,t)$

$\frac{∂U}{∂t}≈\frac{U^{n+1}_i−U^n_i}{dt}$

$\frac{∂U}{∂x}≈\frac{U^{n}_i−U^n_{i-1}}{dx}$

**1) Явная:**

$$\frac{U^{n+1}_i−U^n_i}{dt} + V_i\frac{U^{n}_i−U^n_{i-1}}{dx} = f_i^n$$

$$U_i^{n+1} = U_i^n - \frac{dt}{dx} V_i (U_i^n - U_{i-1}^n) + dt f_i^n$$

**2) Неявная:**

$$\frac{U^{n+1}_i−U^n_i}{dt} + V_i\frac{U^{n+1}_i−U^{n+1}_{i-1}}{dx} = f_i^n$$

$$U_i^{n+1} + \frac{dt}{dx} V_i U_i^{n+1} - \frac{dt}{dx} V_i U_{i-1}^{n+1} = U_i^n + dt f_i^n$$

$$U_i^{n+1} (1 + \frac{dt}{dx} V_i) - \frac{dt}{dx} V_i U_{i-1}^{n+1} = U_i^n + dt f_i^n$$

$$U_i^{n+1} = \frac{\frac{dt}{dx} V_i U_{i-1}^{n+1} + U_i^n + dt f_i^n}{(1 + \frac{dt}{dx} V_i)}$$

```python
# Создаём сетки
for step in dt_dx:
    x = np.arange(x0, x1 + step[1], step[1])
    t = np.arange(t0, t1 + step[0], step[0])
    nx = len(x)
    nt = len(t)
    
    # Функции V(x)
    V = x + 1
    
    # --- Инициализация ---
    U = np.zeros((nt, nx))
    U_analytic = np.zeros((nt, nx))
    U[0, :] = U0(x)          # начальное условие
    U[:, 0] = U_left(t)       # граничное условие
    
    # --- Явная схема ---
    for it in range(0, nt - 1):
        for ix in range(1, nx):
            U[it + 1, ix] = U[it, ix] - (step[0] / step[1]) * V[ix] * (U[it, ix] - U[it, ix - 1]) + step[0] * f
            U_analytic[it + 1, ix] = U_analytic_scalar(t[it], x[ix]) # параллельно считаем значения для аналитического решения
    # --- Профили в разные моменты времени ---
    plt.figure(figsize=(8, 5))
    for n in range(0, nt, nt // 5):  # 5 временных слоёв
        plt.plot(x, U[n, :], label=f"t = {t[n]:.1f}")
    plt.xlabel('x')
    plt.ylabel('U(t, x)')
    plt.title(f'Распределение U по x в разные моменты времени для явной схемы с шагом dt={step[0]} dx={step[1]}')
    plt.legend()
    plt.grid(True)
    plt.show()

    # --- Неявная схема ---
    U_implicit = np.zeros((nt, nx))
    U_implicit[0, :] = U0(x)
    U_implicit[:, 0] = U_left(t)
    
    for it in range(0, nt - 1):
        # Нижнетреугольная система: решение "вперед по x"
        for ix in range(1, nx):
            U_implicit[it + 1, ix] = (U_implicit[it, ix] + step[0] * f + (step[0] / step[1]) * V[ix] * U_implicit[it + 1, ix - 1]) / (1 + (step[0] / step[1]) * V[ix])
    # --- Профили ---
    plt.figure(figsize=(8, 5))
    for n in range(0, nt, nt // 5):
        plt.plot(x, U_implicit[n, :], label=f"t = {t[n]:.1f}")
    plt.xlabel('x')
    plt.ylabel('U(t, x)')
    plt.title(f'Неявная схема: распределение U по x с шагом dt={step[0]} dx={step[1]}')
    plt.legend()
    plt.grid(True)
    plt.show()    
    
    print('Среднеквадратичное отклонение по явной и не явной схемам: ', rmse(U, U_implicit))
    print('Среднеквадратичное отклонение по явной схеме и аналитическому решению: ', rmse(U, U_analytic))
    print('Среднеквадратичное отклонение по не явной схеме и аналитическому решению: ', rmse(U_analytic, U_implicit))
```
