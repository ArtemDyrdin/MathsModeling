```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.linalg import solve
from scipy.interpolate import interp1d
```

```python
U_x0 = lambda x: x * (10.0 - x)**2   # начальное условие U(0,x)
U_0t = lambda t: 0.0                 # граничные условия на x=0 и x=10 (Dirichlet)

D = lambda x: 12.0 - x               # коэффициент диффузии
f = lambda x: 11.0 - x               # коэффициент в реакционной части

x_min, x_max = 0.0, 10.0
t_min, t_max = 0.0, 0.5
```

```python
h_values = [0.5, 0.25, 0.125, 0.0625, 0.03125]

# Оценка: tau <= h^2 / (2*maxD), где maxD = 12.
maxD = 12.0
tau_explicit = [round(0.3 * h*h / (2*maxD), 8) for h in h_values]  # константа 0.3 для безопасности

# Для неявной схемы можно брать более крупные tau
tau_implicit = [0.05, 0.01, 0.005, 0.001, 0.0005]

print("tau (explicit) list:", tau_explicit)
print("tau (implicit) list:", tau_implicit)
```

```python
def explicit_step(u, h, tau):
    N = len(u)-1
    u_new = u.copy()
    for k in range(1,N):
        xk = k*h
        Dk = D_func(xk)
        fk = f_func(xk)
        u_new[k] = u[k] + tau*( Dk*(u[k-1] - 2*u[k] + u[k+1]) / h**2 - fk*u[k] + 5.0 )
    u_new[0] = 0.0; u_new[N] = 0.0
    return u_new
```

```python
solutions_exp = []
meshes_exp = []

solutions_imp = []
meshes_imp = []
```

```python
def explicit_scheme(h, tau):
    x = np.arange(x_min, x_max + h, h)
    t = np.arange(t_min, t_max + tau, tau)
    Nx = len(x)
    Nt = len(t)

    u = np.zeros((Nx, Nt))
    # начальное и граничные условия
    u[:, 0] = U_x0(x)
    u[0, :] = U_0t(t)
    u[-1, :] = U_0t(t)

    for j in range(Nt - 1):
        for i in range(1, Nx - 1):
            xi = x[i]
            Di = D(xi)
            fi = f(xi)
            # FTCS (явная)
            u[i, j+1] = (1 - 2 * Di * tau / (h**2) - fi * tau) * u[i, j] \
                        + (Di * tau / (h**2)) * u[i+1, j] \
                        + (Di * tau / (h**2)) * u[i-1, j] \
                        + 5.0 * tau

        u[0, j+1] = 0.0
        u[-1, j+1] = 0.0

    solutions_exp.append(u[:, -1])
    meshes_exp.append(x)

    print(f"[explicit] h = {h}, tau = {tau}, Nx = {Nx}, Nt = {Nt}")

    # 3D-график (x,t,u)
    X, T = np.meshgrid(x, t)
    Uplot = u.T
    fig = plt.figure(figsize=(12,6))
    ax = fig.add_subplot(111, projection='3d')
    surf = ax.plot_surface(X, T, Uplot, cmap='viridis', edgecolor='none')
    ax.set_xlabel('x'); ax.set_ylabel('t'); ax.set_zlabel('u(x,t)')
    fig.colorbar(surf, shrink=0.5, aspect=5, label='u(x,t)')
    plt.title(f'Explicit scheme: h={h}, tau={tau}')
    plt.show()

    # Профили в несколько моментов времени
    t_vals = np.linspace(0, t_max, 8)
    plt.figure(figsize=(10,6))
    for tt in t_vals:
        j = np.argmin(np.abs(t - tt))
        plt.plot(x, u[:, j], label=f"t={t[j]:.3f}")
    plt.xlabel("x"); plt.ylabel("u(t,x)")
    plt.title("Explicit: профили по x в разные моменты")
    plt.legend(); plt.grid(True)
    plt.show()
```

```python
def implicit_scheme(h, tau):
    x = np.arange(x_min, x_max + h, h)
    t = np.arange(t_min, t_max + tau, tau)
    Nx = len(x)
    Nt = len(t)

    u = np.zeros((Nx, Nt))
    u[:, 0] = U_x0(x)
    u[0, :] = U_0t(t)
    u[-1, :] = U_0t(t)

    for j in range(Nt - 1):
        # матрица A размера (Nx-2) x (Nx-2) для неизвестных внутренних узлов
        A = np.zeros((Nx-2, Nx-2))
        b = np.zeros(Nx-2)

        for i in range(1, Nx-1):
            row = i - 1
            xi = x[i]
            Di = D(xi)
            fi = f(xi)
            dx2 = h*h

            A[row, row] = 1.0 + 2.0 * Di * tau / dx2 + tau * fi
            if row - 1 >= 0:
                A[row, row - 1] = - Di * tau / dx2
            if row + 1 <= Nx-3:
                A[row, row + 1] = - Di * tau / dx2

            b[row] = u[i, j] + tau * 5.0

        u_internal = solve(A, b)
        u[1:-1, j+1] = u_internal
        u[0, j+1] = 0.0
        u[-1, j+1] = 0.0

    solutions_imp.append(u[:, -1])
    meshes_imp.append(x)

    print(f"[implicit] h = {h}, tau = {tau}, Nx = {Nx}, Nt = {Nt}")

    # Графики
    X, T = np.meshgrid(x, t)
    Uplot = u.T
    fig = plt.figure(figsize=(12,6))
    ax = fig.add_subplot(111, projection='3d')
    surf = ax.plot_surface(X, T, Uplot, cmap='viridis', edgecolor='none')
    ax.set_xlabel('x'); ax.set_ylabel('t'); ax.set_zlabel('u(x,t)')
    fig.colorbar(surf, shrink=0.5, aspect=5, label='u(x,t)')
    plt.title(f'Implicit scheme: h={h}, tau={tau}')
    plt.show()

    t_vals = np.linspace(0, t_max, 8)
    plt.figure(figsize=(10,6))
    for tt in t_vals:
        j = np.argmin(np.abs(t - tt))
        plt.plot(x, u[:, j], label=f"t={t[j]:.3f}")
    plt.xlabel("x"); plt.ylabel("u(t,x)")
    plt.title("Implicit: профили по x в разные моменты")
    plt.legend(); plt.grid(True)
    plt.show()
```

```python
for h, tau in zip(h_values, tau_explicit):
    explicit_scheme(h, tau)
```

```python
# Найдём "тонкую" сетку (последний результат) как эталон для сравнения
u_fine_exp = solutions_exp[-1]
x_fine_exp = meshes_exp[-1]
```

```python
# Оценка порядка для явной схемы (L2-норма)
errors_explicit = []
for k in range(len(h_values)-1):
    x_coarse = meshes_exp[k]
    u_coarse = solutions_exp[k]
    interp = interp1d(x_fine_exp, u_fine_exp, kind='cubic', fill_value="extrapolate", bounds_error=False)
    u_fine_interp = interp(x_coarse)
    err = np.sqrt(np.mean((u_coarse - u_fine_interp)**2))
    errors_explicit.append(err)
```

```python
h_plot_exp = h_values[:-1]
plt.figure(figsize=(8,5))
plt.loglog(h_plot_exp, errors_explicit, 'o-')
plt.xlabel('h'); plt.ylabel('L2 error'); plt.title('Explicit: сходимость (L2)'); plt.grid(True, which="both", ls="--")
plt.show()

p_fit_exp = np.polyfit(np.log(h_plot_exp), np.log(errors_explicit), 1)[0]
print(f"Явная схема: приближённый порядок аппроксимации p ≈ {p_fit_exp:.2f}")
```

```python
for h, tau in zip(h_values, tau_implicit):
    implicit_scheme(h, tau)

u_fine_imp = solutions_imp[-1]
x_fine_imp = meshes_imp[-1]
```

```python
# Оценка порядка для неявной схемы (L2-норма)
errors_implicit = []
for k in range(len(h_values)-1):
    x_coarse = meshes_imp[k]
    u_coarse = solutions_imp[k]
    interp = interp1d(x_fine_imp, u_fine_imp, kind='cubic', fill_value="extrapolate", bounds_error=False)
    u_fine_interp = interp(x_coarse)
    err = np.sqrt(np.mean((u_coarse - u_fine_interp)**2))
    errors_implicit.append(err)
```

```python
h_plot_imp = h_values[:-1]
plt.figure(figsize=(8,5))
plt.loglog(h_plot_imp, errors_implicit, 'o-')
plt.xlabel('h'); plt.ylabel('L2 error'); plt.title('Implicit: сходимость (L2)'); plt.grid(True, which="both", ls="--")
plt.show()

p_fit_imp = np.polyfit(np.log(h_plot_imp), np.log(errors_implicit), 1)[0]
print(f"Неявная схема: приближённый порядок аппроксимации p ≈ {p_fit_imp:.2f}")
```

```python
def compute_errors_and_orders(h_values, solutions, meshes, scheme_name="Scheme"):
    errors = []      # ошибки каждой пары сеток
    orders = []      # приближённый порядок

    # Считаем ошибки между каждой парой соседних сеток
    for k in range(len(h_values)-1):
        u_coarse = solutions[k]
        x_coarse = meshes[k]
        u_fine = solutions[k+1]
        x_fine = meshes[k+1]

        # Интерполируем более тонкую сетку на узлы грубой
        interp = interp1d(x_fine, u_fine, kind='cubic', fill_value="extrapolate", bounds_error=False)
        u_fine_interp = interp(x_coarse)

        # L2 ошибка
        err = np.sqrt(np.mean((u_coarse - u_fine_interp)**2))
        errors.append(err)

        # Порядок аппроксимации между этой и предыдущей парой ошибок
        if k == 0:
            orders.append(np.nan)  # для первой пары нет предыдущей ошибки
        else:
            p = np.log(errors[-1] / errors[-2]) / np.log(h_values[k] / h_values[k-1])
            orders.append(p)

    # Вывод результатов
    print(f"\n{scheme_name} - последовательная оценка аппроксимации:")
    for k in range(len(errors)):
        print(f"h = {h_values[k]} -> {h_values[k+1]}, L2 ошибка = {errors[k]:.6e}, порядок ≈ {orders[k]}")

    return errors, orders
```

```python
# Пример построения log-log графика для двух схем
errors_exp, orders_exp = compute_errors_and_orders(h_values, solutions_exp, meshes_exp, "Explicit scheme")
errors_imp, orders_imp = compute_errors_and_orders(h_values, solutions_imp, meshes_imp, "Implicit scheme")

# Лог-лог график ошибок обеих схем
plt.figure(figsize=(8,5))
plt.loglog(h_values[:-1], errors_exp, 'o-', label="Explicit")
plt.loglog(h_values[:-1], errors_imp, 's-', label="Implicit")
plt.xlabel("h")
plt.ylabel("L2 error")
plt.title("Последовательная оценка ошибок: явная vs неявная")
plt.grid(True, which="both", ls="--")
plt.legend()
plt.show()

p_fit_exp2 = np.polyfit(np.log(h_values[:-1]), np.log(errors_exp), 1)[0]
print(f"Явная схема: приближённый порядок аппроксимации p ≈ {p_fit_exp2:.2f}")

p_fit_imp2 = np.polyfit(np.log(h_values[:-1]), np.log(errors_imp), 1)[0]
print(f"Неявная схема: приближённый порядок аппроксимации p ≈ {p_fit_imp2:.2f}")
```
